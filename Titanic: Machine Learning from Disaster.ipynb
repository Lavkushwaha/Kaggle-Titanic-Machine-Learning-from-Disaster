```python
#preprocessing dataframes
#removing the columns with mostly unique items
#Name,Ticket,PassengerId,Cabin
#train_df.Sex[train_df.Sex == 'female']=2 #male 1 , female 2
#S=1,C=2,Q=3
#train_df.Embarked[train_df.Embarked=='Q']=3

#1st try with all ages if accuracy is less than 'NaN' will be replace with 0


train_df=train_df.drop('Name',axis=1)
train_df=train_df.drop('Ticket',axis=1)
train_df=train_df.drop('PassengerId',axis=1)
train_df=train_df.drop('Cabin',axis=1)

train_df.Sex[train_df.Sex == 'female']=2 #male 1 , female 2
train_df.Sex[train_df.Sex == 'male']=1 #male 1 , female 2

train_df.Embarked[train_df.Embarked=='S']=1
train_df.Embarked[train_df.Embarked=='C']=2
train_df.Embarked[train_df.Embarked=='Q']=3

#Test

test_df=test_df.drop('Name',axis=1)
test_df=test_df.drop('Ticket',axis=1)

#for last result csv
list_passenger=test_df['PassengerId']
#for last result csv

test_df=test_df.drop('PassengerId',axis=1)
test_df=test_df.drop('Cabin',axis=1)

test_df.Sex[test_df.Sex == 'female']=2 #male 1 , female 2
test_df.Sex[test_df.Sex == 'male']=1 #male 1 , female 2

test_df.Embarked[test_df.Embarked=='S']=1
test_df.Embarked[test_df.Embarked=='C']=2
test_df.Embarked[test_df.Embarked=='Q']=3

```

```python
#normalise the age and fare beacuse its size is too much for dtype float32
from sklearn import preprocessing

#train data
x = train_df.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
train_df = pd.DataFrame(x_scaled)

#test data

y = test_df.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
y_scaled = min_max_scaler.fit_transform(y)
test_df = pd.DataFrame(y_scaled)

test_df.head(10)
```

```python
test_df.shape,train_df.shape
```

```python
#Convert to numpy Array
train_data= np.array(train_df,dtype='float32')
test_data= np.array(test_df,dtype='float32')

#split features and labels

x_train=train_data[:,1:]
y_train=train_data[:,0]

x_test=test_data[:]



```

```python
x_train.shape,x_test.shape
```

```python
#nun checking training data
np.isnan(x_train).any()  #return true if Nun is found

#nun is replaced with median averaga
x_train[np.isnan(x_train)] = np.median(x_train[~np.isnan(x_train)])

#nun checking testing data
np.isnan(x_test).any()  #return true if Nun is found

#nun is replaced with median averaga
x_test[np.isnan(x_test)] = np.median(x_test[~np.isnan(x_test)])

```

```python
#Model random forest

from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier(n_estimators=100,oob_score=True,max_features=7)

random_forest.fit(x_train, y_train)

y_pred = random_forest.predict(x_test)

random_forest.score(x_train, y_train)
```
